```python
import os
import time
import math
import threading
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Scikit-learn imports
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, RobustScaler, StandardScaler, LabelEncoder, label_binarize
from sklearn.impute import SimpleImputer
from sklearn.metrics import (
    accuracy_score, precision_recall_fscore_support, classification_report,
    confusion_matrix, roc_curve, auc, roc_auc_score, balanced_accuracy_score,
)
from sklearn.metrics import matthews_corrcoef

# Imblearn for ADASYN + Tomek
from imblearn.over_sampling import ADASYN, RandomOverSampler
from imblearn.under_sampling import TomekLinks

# TensorFlow and Keras
import tensorflow as tf
from tensorflow.keras import layers, models

# Psutil for RAM tracking
try:
    import psutil
    _PSUTIL_AVAILABLE = True
except ImportError:
    _PSUTIL_AVAILABLE = False

# ================== CONFIGURATION ==================
ALL_DATA = Path(r"C:\Users\deepa\OneDrive\Desktop\Research\GNN\ADASYN+Tomek Links\all_three.csv")
OUT_DIR = Path("./outputs_adasyn_tomek"); OUT_DIR.mkdir(parents=True, exist_ok=True)

# One of: "decimal", "minmax", "robust", "standard"
SCALER_NAME = "robust"

TARGET_COL = "label"
SEED = 42
EPOCHS = 10
BATCH_SIZE = 128
TRAIN_RATIOS = [0.6, 0.7, 0.8]

np.random.seed(SEED)
tf.random.set_seed(SEED)

# ================== GPU CHECK (optional) ==================
print("TF version:", tf.__version__)
print("GPUs:", tf.config.list_physical_devices('GPU'))
_gpus = tf.config.list_physical_devices('GPU')
if _gpus:
    try:
        for g in _gpus: tf.config.experimental.set_memory_growth(g, True)
        print("Enabled GPU memory growth.")
    except Exception as e:
        print("Could not set memory growth:", e)

# ================== SCALERS & HELPERS ==================
class DecimalScaler:
    """Scale each feature x_j by 10^ceil(log10(max|x_j|)) -> (-1,1)."""
    def fit(self, X):
        m = np.maximum(np.max(np.abs(X), axis=0), 1e-12)
        self.j_ = np.ceil(np.log10(m)).astype(np.float32)
        return self
    def transform(self, X):
        denom = (10.0 ** self.j_)
        return X / denom
    def fit_transform(self, X):
        return self.fit(X).transform(X)

def get_scaler(name):
    nm = name.lower()
    if nm == 'decimal': return DecimalScaler()
    if nm == 'minmax': return MinMaxScaler()
    if nm == 'robust': return RobustScaler(quantile_range=(25.0, 75.0))
    if nm == 'standard': return StandardScaler()
    raise ValueError('Unknown scaler: ' + str(name))

def apply_adasyn_tomek_safe(X, y):
    """
    Train-only resampling:
      1) ADASYN to synthesize minority samples (safe fallback to ROS if needed)
      2) TomekLinks to clean borderline samples (under-sampling)
    """
    note_parts = []
    try:
        counts = pd.Series(y).value_counts()
        if counts.min() < 6:
            ros = RandomOverSampler(random_state=SEED)
            X_res, y_res = ros.fit_resample(X, y)
            note_parts.append(f'ROS(fallback, min_class={counts.min()})')
        else:
            adasyn = ADASYN(random_state=SEED)
            X_res, y_res = adasyn.fit_resample(X, y)
            note_parts.append('ADASYN')
    except Exception as e:
        ros = RandomOverSampler(random_state=SEED)
        X_res, y_res = ros.fit_resample(X, y)
        note_parts.append(f'ROS(fallback:{str(e)})')

    # TomekLinks cleaning (works on majority class pairs)
    try:
        tl = TomekLinks()  # default: 'auto' remove='auto'
        X_clean, y_clean = tl.fit_resample(X_res, y_res)
        note_parts.append('TomekLinks')
        return (X_clean, y_clean), ' + '.join(note_parts)
    except Exception as e:
        note_parts.append(f'TomekLinks(skip:{str(e)})')
        return (X_res, y_res), ' + '.join(note_parts)

# RAM Monitoring Class
class PeakRSS:
    def __enter__(self):
        self.max_rss = None; self._stop = False
        if _PSUTIL_AVAILABLE:
            self._proc = psutil.Process(os.getpid())
            self.max_rss = self._proc.memory_info().rss
            def _poll():
                while not self._stop:
                    try:
                        rss = self._proc.memory_info().rss
                        if rss is not None and (self.max_rss is None or rss > self.max_rss): self.max_rss = rss
                    except Exception: pass
                    time.sleep(0.2)
            self._t = threading.Thread(target=_poll, daemon=True); self._t.start()
        return self
    def __exit__(self, exc_type, exc, tb):
        if _PSUTIL_AVAILABLE:
            self._stop = True
            try: self._t.join(timeout=1.0)
            except Exception: pass

# ================== PLOTTING ==================
def plot_confusion_matrix(cm, class_names, outpath, title="Confusion Matrix"):
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
    plt.title(title); plt.xlabel("Predicted"); plt.ylabel("True"); plt.tight_layout()
    plt.savefig(outpath)
    plt.show()

def plot_roc(y_true, y_probs, class_names, outpath, title="ROC Curve"):
    try:
        y_true_bin = label_binarize(y_true, classes=np.arange(len(class_names)))
        plt.figure(figsize=(8, 6))
        for i, c in enumerate(class_names):
            fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_probs[:, i])
            auc_score = auc(fpr, tpr)
            plt.plot(fpr, tpr, label=f"Class {c} (AUC = {auc_score:.2f})")
        plt.plot([0, 1], [0, 1], "k--", label="Chance")
        plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
        plt.title(title); plt.legend(loc="lower right"); plt.tight_layout()
        plt.savefig(outpath)
        plt.show()
    except Exception as e:
        print("ROC plotting failed:", e)

# ================== SIMPLE GNN LAYERS (fixed) ==================
def make_feature_chain_adj(n_features: int):
    A = np.zeros((n_features, n_features), dtype=np.float32)
    for i in range(n_features):
        A[i, i] = 1.0
        if i - 1 >= 0: A[i, i-1] = 1.0
        if i + 1 < n_features: A[i, i+1] = 1.0
    d = np.sum(A, axis=1)
    d_inv_sqrt = 1.0 / np.sqrt(np.maximum(d, 1e-12))
    D_inv_sqrt = np.diag(d_inv_sqrt.astype(np.float32))
    A_hat = D_inv_sqrt @ A @ D_inv_sqrt
    return A_hat.astype(np.float32)

class GraphConv(layers.Layer):
    """H = act(Â X W + b) where Â is fixed normalized adjacency (features-as-nodes)."""
    def __init__(self, A_hat: np.ndarray, units: int, activation="relu", **kwargs):
        super().__init__(**kwargs)
        self.units = int(units)
        self.activation = tf.keras.activations.get(activation)
        self._A_hat_np = A_hat

    def build(self, input_shape):
        in_ch = int(input_shape[-1])
        dtype_ = self.dtype if self.dtype is not None else tf.float32
        self.A_hat = tf.constant(self._A_hat_np, dtype=dtype_)
        self.kernel = self.add_weight(
            name="kernel", shape=(in_ch, self.units),
            dtype=dtype_, initializer="glorot_uniform", trainable=True
        )
        self.bias = self.add_weight(
            name="bias", shape=(self.units,),
            dtype=dtype_, initializer="zeros", trainable=True
        )

    def call(self, X):
        # X: (b, n, in_ch)
        H = tf.einsum('ij,bjk->bik', self.A_hat, X)   # (b, n, in_ch)
        H = tf.matmul(H, self.kernel)                 # (b, n, units)
        H = tf.nn.bias_add(H, self.bias)
        return self.activation(H)

def build_gnn_single(n_features: int, num_classes: int):
    A_hat = make_feature_chain_adj(n_features)
    f32 = tf.float32
    inp = layers.Input(shape=(n_features, 1), dtype=f32)
    x = GraphConv(A_hat, units=64, activation="relu", dtype=f32)(inp)
    x = layers.Dropout(0.15)(x)
    x = GraphConv(A_hat, units=64, activation="relu", dtype=f32)(x)
    x = layers.Dropout(0.15)(x)
    x = layers.GlobalAveragePooling1D()(x)
    if num_classes > 2:
        out = layers.Dense(num_classes, activation="softmax", dtype=f32)(x)
        loss = "sparse_categorical_crossentropy"
    else:
        out = layers.Dense(1, activation="sigmoid", dtype=f32)(x)
        loss = "binary_crossentropy"
    model = models.Model(inp, out)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
                  loss=loss, metrics=["accuracy"])
    return model

# ================== BIAS METRICS ==================
def compute_bias_metrics(y_true, y_pred, probs, classes):
    """
    Returns:
      summary: dict of global metrics & disparity indicators
      per_class_df: per-class precision/recall/f1/support and AUC
    """
    n_classes = len(classes)
    # Per-class metrics
    prec_i, rec_i, f1_i, support_i = precision_recall_fscore_support(
        y_true, y_pred, labels=classes, average=None, zero_division=0
    )
    # AUC per class (one-vs-rest) if probs available
    auc_i = []
    try:
        y_true_bin = label_binarize(y_true, classes=classes)
        for i in range(n_classes):
            fpr, tpr, _ = roc_curve(y_true_bin[:, i], probs[:, i])
            auc_i.append(auc(fpr, tpr))
    except Exception:
        auc_i = [np.nan] * n_classes

    per_class_df = pd.DataFrame({
        'class': classes,
        'support': support_i,
        'precision': prec_i,
        'recall': rec_i,
        'f1': f1_i,
        'auc_ovr': auc_i
    })

    # Global metrics
    acc = accuracy_score(y_true, y_pred)
    bal_acc = balanced_accuracy_score(y_true, y_pred)
    macro_p, macro_r, macro_f1, _ = precision_recall_fscore_support(
        y_true, y_pred, average='macro', zero_division=0
    )
    weighted_p, weighted_r, weighted_f1, _ = precision_recall_fscore_support(
        y_true, y_pred, average='weighted', zero_division=0
    )
    try:
        roc_auc_global = roc_auc_score(y_true, probs, multi_class="ovr" if n_classes > 2 else "raise")
    except Exception:
        roc_auc_global = float("nan")

    # Disparities (max - min across classes)
    prec_disp = float(np.nanmax(prec_i) - np.nanmin(prec_i))
    rec_disp  = float(np.nanmax(rec_i)  - np.nanmin(rec_i))
    f1_disp   = float(np.nanmax(f1_i)   - np.nanmin(f1_i))

    # G-mean of recalls (robust to imbalance)
    try:
        gmean = float(np.prod(np.maximum(rec_i, 1e-12)) ** (1.0 / len(rec_i)))
    except Exception:
        gmean = float("nan")

    # MCC
    try:
        mcc = matthews_corrcoef(y_true, y_pred)
    except Exception:
        mcc = float("nan")

    summary = {
        "Accuracy": acc,
        "Balanced_Accuracy": bal_acc,
        "Macro_Precision": macro_p,
        "Macro_Recall": macro_r,
        "Macro_F1": macro_f1,
        "Weighted_Precision": weighted_p,
        "Weighted_Recall": weighted_r,
        "Weighted_F1": weighted_f1,
        "ROC_AUC_OVR": roc_auc_global,
        "Precision_Disparity(max-min)": prec_disp,
        "Recall_Disparity(max-min)": rec_disp,
        "F1_Disparity(max-min)": f1_disp,
        "GMean_Recall": gmean,
        "MCC": mcc
    }
    return summary, per_class_df

# ================== 1. LOAD & PREPROCESS DATA ==================
df = pd.read_csv(ALL_DATA)
df = df.dropna(subset=[TARGET_COL])

X_features = df.drop(columns=[TARGET_COL]).apply(pd.to_numeric, errors='coerce')
y_labels = df[TARGET_COL].values

imputer = SimpleImputer(strategy='median')
X_imputed = imputer.fit_transform(X_features)

le = LabelEncoder()
y_encoded = le.fit_transform(y_labels)
classes = np.unique(y_encoded)
class_names = [str(c) for c in classes]
print(f"Running for {SCALER_NAME.upper()} Scaler\nClasses: {classes}")

# ================== 2. RUN EXPERIMENTS FOR EACH RATIO ==================
overall_results = []
for ratio in TRAIN_RATIOS:
    print(f"\n{'='*20} Train split {int(ratio*100)}% {'='*20}")

    # --- Split & scale ---
    X_tr, X_te, y_tr, y_te = train_test_split(
        X_imputed, y_encoded, train_size=ratio, stratify=y_encoded, random_state=SEED
    )
    scaler = get_scaler(SCALER_NAME)
    X_tr_s = scaler.fit_transform(X_tr)
    X_te_s = scaler.transform(X_te)

    # --- Class counts (before) ---
    before_counts = pd.Series(y_tr).value_counts().sort_index()
    print("Class counts (train BEFORE):")
    print(before_counts.to_string())

    # --- ADASYN + TomekLinks (train only) ---
    (X_tr_res, y_tr_res), resample_note = apply_adasyn_tomek_safe(X_tr_s, y_tr)
    after_counts = pd.Series(y_tr_res).value_counts().sort_index()
    print(f"Resampling: {resample_note}")
    print("Class counts (train AFTER):")
    print(after_counts.to_string())

    # --- Reshape for Single-GNN ---
    n_feats = X_tr_res.shape[1]
    X_train_model = X_tr_res.astype(np.float32).reshape((-1, n_feats, 1))
    X_test_model  = X_te_s.astype(np.float32).reshape((-1, n_feats, 1))

    # --- Build/Train/Predict ---
    device = '/GPU:0' if tf.config.list_physical_devices('GPU') else '/CPU:0'
    with tf.device(device):
        model = build_gnn_single(n_features=n_feats, num_classes=len(classes))
        start_time = time.perf_counter()
        with PeakRSS() as mem_monitor:
            model.fit(
                X_train_model, y_tr_res,
                validation_split=0.15, epochs=EPOCHS, batch_size=BATCH_SIZE, verbose=1
            )
            if len(classes) > 2:
                probs = model.predict(X_test_model, verbose=0)
                preds = np.argmax(probs, axis=1)
            else:
                probs_bin = model.predict(X_test_model, verbose=0).ravel()
                probs = np.stack([1.0 - probs_bin, probs_bin], axis=1)
                preds = (probs_bin >= 0.5).astype(int)
        runtime_sec = time.perf_counter() - start_time
        ram_used_mb = (mem_monitor.max_rss / (1024**2)) if (_PSUTIL_AVAILABLE and mem_monitor.max_rss is not None) else float("nan")

    # --- Confusion Matrix & ROC ---
    cm = confusion_matrix(y_te, preds)
    cm_path  = OUT_DIR / f'cm_{SCALER_NAME}_split_{int(ratio*100)}.png'
    roc_path = OUT_DIR / f'roc_{SCALER_NAME}_split_{int(ratio*100)}.png'
    plot_confusion_matrix(cm, class_names, cm_path, f'CM ({SCALER_NAME}, Train Ratio: {ratio})')
    plot_roc(y_te, probs, class_names, roc_path, f'ROC ({SCALER_NAME}, Train Ratio: {ratio})')

    # --- Standard metrics ---
    accuracy = accuracy_score(y_te, preds)
    precision, recall, f1, _ = precision_recall_fscore_support(y_te, preds, average='weighted', zero_division=0)
    try:
        roc_auc = roc_auc_score(y_te, probs, multi_class="ovr" if len(classes) > 2 else "raise")
    except Exception:
        roc_auc = float("nan")

    # --- Bias diagnostics ---
    bias_summary, per_class_df = compute_bias_metrics(y_te, preds, probs, classes)
    # Save per-class metrics for this split
    per_class_path = OUT_DIR / f'per_class_metrics_{SCALER_NAME}_split_{int(ratio*100)}.csv'
    per_class_df.to_csv(per_class_path, index=False)

    # --- Print bias diagnostics (quick view) ---
    print("\nBias diagnostics (summary):")
    for k, v in bias_summary.items():
        if isinstance(v, float):
            print(f"{k}: {v:.4f}")
        else:
            print(f"{k}: {v}")

    # --- Collect overall summary row ---
    row = {
        'Scaler': SCALER_NAME,
        'Train Ratio': ratio,
        'Resampling': resample_note,
        'Accuracy': accuracy,
        'Precision(weighted)': precision,
        'Recall(weighted)': recall,
        'F1(weighted)': f1,
        'ROC_AUC(OVR)': roc_auc,
        'Balanced_Accuracy': bias_summary['Balanced_Accuracy'],
        'Macro_F1': bias_summary['Macro_F1'],
        'Precision_Disparity': bias_summary['Precision_Disparity(max-min)'],
        'Recall_Disparity': bias_summary['Recall_Disparity(max-min)'],
        'F1_Disparity': bias_summary['F1_Disparity(max-min)'],
        'GMean_Recall': bias_summary['GMean_Recall'],
        'MCC': bias_summary['MCC'],
        'Runtime(s)': runtime_sec,
        'Peak_RAM(MB)': ram_used_mb
    }
    overall_results.append(row)

    # --- Human-readable report ---
    print(f"\nRESULTS ({SCALER_NAME}, split {int(ratio*100)}%):")
    print("-" * 50)
    print(f"Accuracy: {accuracy:.4f} | Precision(w): {precision:.4f} | Recall(w): {recall:.4f} | F1(w): {f1:.4f} | ROC-AUC: {roc_auc:.4f}")
    print(f"Balanced Acc: {bias_summary['Balanced_Accuracy']:.4f} | Macro F1: {bias_summary['Macro_F1']:.4f} | G-Mean: {bias_summary['GMean_Recall']:.4f} | MCC: {bias_summary['MCC']:.4f}")
    print(f"Dispersion → Precision: {bias_summary['Precision_Disparity(max-min)']:.4f} | Recall: {bias_summary['Recall_Disparity(max-min)']:.4f} | F1: {bias_summary['F1_Disparity(max-min)']:.4f}")
    print("\nClassification Report:")
    print(classification_report(y_te, preds, digits=4))

# ================== 3. SAVE FINAL SUMMARY ==================
summary_df = pd.DataFrame(overall_results)
summary_df_path = OUT_DIR / f'metrics_summary_{SCALER_NAME}_ADASYN_Tomek.csv'
summary_df.to_csv(summary_df_path, index=False)
print(f"\n✅ Done. Summary saved to: {summary_df_path}")
print(summary_df.to_string(index=False, float_format=lambda x: f'{x:.4f}'))
```
